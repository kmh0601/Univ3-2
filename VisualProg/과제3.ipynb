```python
import random
import time

def compareTyping(a, b):
    c = [" "," "]
    for i in range(len(a)):
        if len(b)-1 < i: # 입력받은 타이핑이 작으면
            c.append("^")
        elif a[i] != b[i]: # 잘못 입력했으면
            c.append("^")
        else:
            c.append(" ")
            
    if len(b) > len(a): # 넘치게 입력하면 모두 오타로 간주
        for i in range(len(b)-len(a)):
            c.append("^")
            
    return c

typingFile = open("./littleprince.txt","r")
typingList = typingFile.readlines()
typingFile.close()
randIndex = random.randint(0,len(typingList))

for i in range(5):
    temp = random.randint(0,len(typingList))
    while randIndex == temp:
        temp = random.randint(0,len(typingList))
    randIndex = temp
    
    outputTyping = typingList[randIndex]
    print("({}/5) 다음을 입력하세요".format(i+1))
    print("=",outputTyping)
    
    start = time.time() # 시간측정 시작
    inputTyping = input("> ")
    end = time.time() # 시간측정 끝
    result = compareTyping(outputTyping[:-1], inputTyping) # [:-1] -> 개행문자 제거
    
    if "^" in result:
        for j in result:
            print(j, end = "")
        print()
    
    speed = len(inputTyping)-result.count("^")
    if speed > 0:
        speed = (speed / (end-start)) * 60
    else:
        speed = 0
    print("({} strokes/minute)".format(int(speed)))
```

```python
import random
ini = 0
def isCorret(word, newWord, size):
    if word[-1] == newWord[0] and len(newWord)-newWord.count(" ") == size: # 사전에 띄어쓰기가 포함된 단어가 있음
        return True
    else:
        return False
    
def isNotUsed(word):
    global usedWordList
    if word in usedWordList:
        return False
    else:
        return True

dictFile = open('./dict_test/dict_test_utf8.txt','r')
dictInfo = dictFile.readlines()
dictFile.close()
dictList = {}
for i in dictInfo:
    data = i.split(" : ")
    if len(data) == 2:
        dictList[data[0]] = data[1]
    elif len(data) == 1:
        dictList[data[0]] = None
wordList = list(dictList.keys())
usedWordList = []
comWin = False
userWin = False
gameRound = 1
score = 0

comSize = random.randint(3,10)
comWord = random.choice(wordList)
print(gameRound, "round")
print("컴퓨터 : {} {}".format(comWord, comSize))

while True:
    #user turn
    while True:
        userInput = list(input("{}글자의 단어와 글자수를 입력하세요.".format(comSize)).split(" "))
        if len(userInput) == 2:
            break
        else:
            print("단어와 글자수를 입력해야 합니다!")
            continue
    userWord = userInput[0]
    userSize = int(userInput[1])
    if isCorret(comWord, userWord, comSize) and isNotUsed(userWord) and (userWord in wordList):
        usedWordList.append(userWord)
        score += 1
        print("1점 획득. 현재 점수%d"%(score))
        comWin = False
    else:
        comWin = True
    
    if comWin:
        break
    
    gameRound += 1
    # com turn
    size = random.randint(3,10)
    for i in wordList:
        if isCorret(userWord, i, userSize) and isNotUsed(i):
            comWord = i
            userWin = False
            break
        else:
            userWin = True
    if userWin:
        break   
        
    comSize = random.randint(3,10)
    usedWordList.append(comWord)
    print(gameRound,"round")
    print("컴퓨터 : {} {}".format(comWord, comSize))

if comWin:
    print("컴퓨터가 이겼습니다.")
elif userWin:
    print("사용자가 이겼습니다.")
print("최종 점수는 %d점 입니다."%score)
```

```python
covidCSV = open("./owid-covid-data.csv","r")
col = covidCSV.readline().split(",")

data = covidCSV.readlines()
covidCSV.close()
covid_data = []
for line in data:
    covid_data.append(list(line.split(",")))
    
while True:
    #국가 정하기
    userInput = input("나라 이름을 입력하세요. ").lower()
    locations = []
    for line in covid_data: # 비교할때 데이터 리스트와 사용자 입력의 모두 나라 이름을 소문자로 바꿔서 비교함
        if (userInput in line[col.index("location")].lower()) and (line[col.index("location")] not in locations):
            locations.append(line[col.index("location")])

    #국가 확인
    if len(locations) == 0:
        print("찾으시는 국가가 없습니다.")
        continue
    elif len(locations) == 1:
        yesOrNo = input("찾으시는 국가가 '{}' 맞습니까? (yes/no)".format(locations[0]))
        if yesOrNo == "yes":
            location = locations[0]
        else:
            continue
    else:
        for i in locations:
            print(i)
        location = ""
        while location not in locations:
            location = input("제시한 국가 중 선택해주세요. > ")
    
    #최신 데이터 행에 접근
    for i in range(len(covid_data)):
        if covid_data[i][col.index("location")] == location and covid_data[i+1][col.index("location")] != location:
            index = i
            break
            
            
    ### _데이터 종류마다 데이터가 존재하는 가장 최신 날짜가 달라서 따로 접근함_
    #최신 데이터 행으로부터 '새로운 확진자' 데이터가 존재하는 가장 최신 행 접근
    caseIndex = index
    while True:
        if covid_data[caseIndex][col.index("new_cases")] != '' and covid_data[caseIndex][col.index("new_cases")] != '0.0':
            break
        else:
            caseIndex -= 1
    
    print("날짜 :",covid_data[caseIndex][col.index("date")])
    print("총 확진자 수 :",covid_data[caseIndex][col.index("total_cases")])
    print("새로운 확진자 수 :" ,covid_data[caseIndex][col.index("new_cases")])
    print()
    
    #최신 데이터 행으로부터 '새로운 사망자' 데이터가 존재하는 가장 최신 행 접근
    deathsIndex = index
    while True:
        if covid_data[deathsIndex][col.index("new_deaths")] != '' and covid_data[deathsIndex][col.index("new_deaths")] != '0.0':
            break
        else:
            deathsIndex -= 1
    
    print("날짜 :",covid_data[deathsIndex][col.index("date")])
    print("총 사망자 수 :",covid_data[deathsIndex][col.index("total_deaths")])
    print("새로운 사망자 수 :",covid_data[deathsIndex][col.index("new_deaths")])
    print()
    
    #최신 데이터 행으로부터 '새로운 백신 접종자' 데이터가 존재하는 가장 최신 행 접근
    vaccinationsIndex = index
    while True:
        if covid_data[vaccinationsIndex][col.index("new_vaccinations")] != '' and covid_data[vaccinationsIndex][col.index("new_vaccinations")] != '0.0':
            break
        else:
            vaccinationsIndex -= 1
    
    print("날짜 :",covid_data[vaccinationsIndex][col.index("date")])
    print("총 백신 접종자 수 :",covid_data[vaccinationsIndex][col.index("total_vaccinations")])
    print("새로운 백신 접종자 수 :",covid_data[vaccinationsIndex][col.index("new_vaccinations")])
    print()
```
